```python
import xarray as xr
import glob
```

```python
files_BAU = glob.glob("/mnt/data-summer-user/data_mod/BAU_V2/GEOSChem.MercuryChem.2015*.nc4")
print(files_BAU)
files_HIST = glob.glob("/mnt/data-summer-user/data_mod/HIST_V2/GEOSChem.MercuryChem.2015*.nc4")
print(files_HIST)

files_BAU_dry = glob.glob("/mnt/data-summer-user/data_mod/BAU_V2/GEOSChem.DryDep.2015*.nc4")
print(files_BAU_dry)
files_HIST_dry = glob.glob("/mnt/data-summer-user/data_mod/HIST_V2/GEOSChem.DryDep.2015*.nc4")
print(files_HIST_dry)


files_BAU_wet = glob.glob("/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/BAU/GEOSChem.WetLossTotal.2015*.nc4")
print(files_BAU_wet)
files_HIST_wet = glob.glob("/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/HIST/GEOSChem.WetLossTotal.2015*.nc4")
print(files_HIST_wet)
```

```python
ds_BAU = xr.open_mfdataset(files_BAU, combine="by_coords")
print(ds_BAU)
ds_HIST = xr.open_mfdataset(files_HIST, combine="by_coords")
print(ds_HIST)

ds_BAU_dry = xr.open_mfdataset(files_BAU_dry, combine="by_coords")
print(ds_BAU_dry)
ds_HIST_dry = xr.open_mfdataset(files_HIST_dry, combine="by_coords")
print(ds_HIST_dry)

ds_BAU_wet = xr.open_mfdataset(files_BAU_wet, combine="by_coords")
print(ds_BAU_wet)
ds_HIST_wet = xr.open_mfdataset(files_HIST_wet, combine="by_coords")
print(ds_HIST_wet)
```

```python
ds_BAU_mean = ds_BAU.mean(dim="time")
output_path_BAU = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/BAU/GEOSChem.MercuryEmis.2015_m.nc4"
ds_BAU_mean.to_netcdf(output_path_BAU)
print(f"Fichier moyen enregistré : {output_path_BAU}")

ds_HIST_mean = ds_HIST.mean(dim="time")
output_path_HIST = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/HIST/GEOSChem.MercuryEmis.2015_m.nc4"
ds_HIST_mean.to_netcdf(output_path_HIST)
print(f"Fichier moyen enregistré : {output_path_HIST}")


ds_BAU_mean_dry = ds_BAU_dry.mean(dim="time")
output_path_BAU_dry = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/BAU/GEOSChem.DryDep.2015_m.nc4"
ds_BAU_mean_dry.to_netcdf(output_path_BAU_dry)
print(f"Fichier moyen enregistré : {output_path_BAU_dry}")

ds_HIST_mean_dry = ds_HIST_dry.mean(dim="time")
output_path_HIST_dry = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/HIST/GEOSChem.DryDep.2015_m.nc4"
ds_HIST_mean_dry.to_netcdf(output_path_HIST_dry)
print(f"Fichier moyen enregistré : {output_path_HIST_dry}")


ds_BAU_mean_wet = ds_BAU_wet.mean(dim="time")
output_path_BAU_wet = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/BAU/GEOSChem.WetLossTotal.2015_m.nc4"
ds_BAU_mean_wet.to_netcdf(output_path_BAU_wet)
print(f"Fichier moyen enregistré : {output_path_BAU_wet}")

ds_HIST_mean_wet = ds_HIST_wet.mean(dim="time")
output_path_HIST_wet = "/home/jovyan/colomb_ige-m2-internship-2025_MERCAMAZON-project/data_mod/HIST/GEOSChem.WetLossTotal.2015_m.nc4"
ds_HIST_mean_wet.to_netcdf(output_path_HIST_wet)
print(f"Fichier moyen enregistré : {output_path_HIST_wet}")

```
